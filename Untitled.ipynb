```python
import geopandas as gpd
```

```python
nodes = gpd.read_file('data/caba-network.gpkg', layer='nodes')
edges = gpd.read_file('data/caba-network.gpkg', layer='edges')
```

```python
nodes = (
    nodes.reset_index()
        .rename(columns={'index': 'start'})
        .rename(columns={'x': 'lon', 'y': 'lat'})
)
# nodes['start'] = nodes['start'] + 1
nodes.head()
```

```python
nodes[['lat', 'lon']].to_csv('nodes-caba.csv', index=False)
```

```python
nodes[['lat', 'lon']].head()
```

```python
from random import random
```

```python
edges_caba = (
    edges[['from', 'to', 'length']].merge(
        nodes[['osmid', 'start']],
        left_on='from',
        right_on='osmid'
    )
    .merge(
        nodes[['osmid', 'start']].rename(columns={'start': 'end'}),
        left_on='to',
        right_on='osmid'
    )
    .drop(columns=['from', 'to', 'osmid_x', 'osmid_y'])
    .rename(columns={'length': 'distance'})
    .loc[:, ['start', 'end', 'distance']]
)
edges_caba['times_by_hour'] = '\t'.join([str(random()) for i in range(24)])
```

```python
edges_caba.head()
```

```python
edges_caba.to_csv('edges-caba.csv', index=False)
```
